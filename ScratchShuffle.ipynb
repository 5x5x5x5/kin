```python
import numpy as np
import pandas as pd
```

### Load dataframe and count number of cases positive and negative cases 

```python
isozyme2c19 = pd.read_csv('data/2c19.csv')
```

```python
isozyme2c19.columns
```

```python
# Renaming the Activity Score column to conform to Python syntax
isozyme2c19.rename(columns={'2c19ActivityScore': 'ActivityScore'}, inplace=True)
```

```python
# Number of substances with an activity scores greater than or equal to 40
n_pos = (isozyme2c19.ActivityScore >= 40).sum()
n_pos
```

```python
# Number of substances with an activity score below 40
n_neg = (isozyme2c19.ActivityScore < 40).sum()
n_neg
```

## First attempt at downsampling the negative cases

```python
# First attempt - method adapted from DataRobot post about scikit-learn classification

# Downsample negative cases -- there are many more negatives than positives

indices = np.where(isozyme2c19.ActivityScore < 40)[0]
rng = np.random.RandomState(50)  # sets seed for random number generator
rng.shuffle(indices)             # different seed numbers result in different shuffle
n_pos = (isozyme2c19.ActivityScore >= 40).sum()
balanced = isozyme2c19.drop(isozyme2c19.index[indices[n_pos:]])

balanced.head()
```

```python
n_pos = (balanced.ActivityScore >= 40).sum()
n_neg = (balanced.ActivityScore < 40).sum()
n_neg, n_pos
```

```python
# Repeat code with different seed number

indices = np.where(isozyme2c19.ActivityScore < 40)[0]
rng = np.random.RandomState(3489453655)  # sets seed for random number generator
rng.shuffle(indices)             # different seed numbers result in different shuffle
n_pos = (isozyme2c19.ActivityScore >= 40).sum()
balanced2 = isozyme2c19.drop(isozyme2c19.index[indices[n_pos:]])

balanced2.head()
```

```python
n_pos = (balanced2.ActivityScore >= 40).sum()
n_neg = (balanced2.ActivityScore < 40).sum()
n_neg, n_pos
```
